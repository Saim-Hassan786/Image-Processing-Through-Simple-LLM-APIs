# Image Processing Through LLM's API

**We can use the LLM models for not only processing the text but we can also use it for the image processing as well. To proceed with this we first have to download the Google Generative AI Models SDK and use the Gemini's ALI key that we have created from Google AI Studio**

```python
!pip install -U -q "google-generativeai>=0.7.2"
import google.generativeai as genai
from google.colab import userdata
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel("gemini-1.5-flash")
```

**Now with this model variable we can access a set of different methods and attributes and amongst them to get the response from LLM, we can make use of its "generate_content" method as follows**

```python
response = model.generate_content("Tell me about Python Language in 2 lines")
print(response.text)
```

**Now as per our assignment to give the model an image to process we first have to import the image in this colab environment by the use of following commands**

```python
!curl -o image.jpg "https://cdn.pixabay.com/photo/2012/11/28/10/34/rocket-launch-67643_1280.jpg"
```

**Now we can also open the image saved earlier by the use of following Library**

```python
import PIL.Image
img = PIL.Image.open("image.jpg")
img
```

**Now as you can the Image is sucessfully uploaded and opened above , now we can also get our image processed from our LLM model as well**

```python
prompt = f"""
Identify the object in this image and tell me what is exactly happenging in this image and also tell me the biggest operational organizations in this object shown in the image
"""
response = model.generate_content([prompt, img])
print(response.text)
```

**Now instead of passing the prompt directly in the "generate_content" we saved our prompt in a variable named prompt and the Image we want to process in the variable named img and at the end asked the LLM model for the completion of our prompt by simply passing our variables 'prompt' and 'img' in it with in the list and passing the list as a whole because list maintains an ordered structure, so the model knows the relationship between inputs (e.g., prompt first, then image).**
